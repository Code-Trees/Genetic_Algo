```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(100,1)
```

```python
y = 5*x +6
```

```python
plt.scatter(x,y)
```

```python
n = len(x)
def Gradient(itr,lr):
    iteration = itr
    learning_rate = lr
    m_curr=c_curr = 0
    loss_val = []
    m_val = []
    c_val = []
    for i in range(iteration):
        y_pred = m_curr*x + c_curr
        m_der = -(2/n)*sum(x*(y -y_pred))
        c_der = -(2/n)*sum(y-y_pred)
        m_curr = m_curr - lr*m_der
        c_curr = c_curr -lr*c_der
        m_val.append(m_curr)
        c_val.append(c_curr)
        loss = (1/n)*sum([val**2 for val in (y-y_pred)])
        loss_val.append(loss)
        print('loss {} ,iter {}'.format(loss,i))
    return loss_val,m_val,c_val,y_pred
```

```python
itr = 500
lrt = 0.065
loss,m_val,c_val,y_pred = Gradient(itr,lrt)
```
